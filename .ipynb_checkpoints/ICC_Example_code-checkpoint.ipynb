```python
#use with environment MRI_Analysis
from pathlib import Path
import nibabel as nib
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
from nilearn import plotting
from nilearn.masking import compute_multi_brain_mask
from pyrelimri import brain_icc, icc, similarity


# ICC toolbox python
#https://pyrelimri.readthedocs.io/en/latest/icc_funcs.html
```

```python
# automatically get path from script
script_path = os.getcwd()
path = Path(script_path)
path
```

```python
# Choose your beta or con images wisely (see Literature folder. 
# BEWARE: betas images from sham and "active" are calculated by a general linear model containing all stages in one beta regressor and the stages
# in a paramtreic modulator (extra regressor or extra betta image). If you have a general linear model containing a regressor for each stage you 
# have to adapt the data loading so that in the sham list all beta images for all stages are loaded and in the active arm the same

sub_root = os.path.join(path,'betas') # 
figure_path = os.path.join(path,'Figures')

# you can use activity mask or any other mask you like the restrict the area of the ICC (mask example for P1 Learning bigger Control p <0.001 uncorrected
activity_mask= os.path.join(path,'masks','mask_bin_P1_LbC.nii')
  
```

```python
# should show correct path to subjects and figure folder
print(sub_root)
print(figure_path)
```

```python
# Shows als Subjects with images
def list_first_level_directories(folder):
    return [d for d in os.listdir(folder) if os.path.isdir(os.path.join(folder, d))]

# Example usage
first_level_dirs = list_first_level_directories(sub_root)
print(sorted(first_level_dirs))
```

```python

def save_images(brain_icc_dict, figure_path,*args):
    for key in list(brain_icc_dict.keys()):   
        nib.save(brain_icc_dict[key], os.path.join(figure_path, f'{key}_{args[0]}.nii'))
```

```python

def save_images_mask(brain_icc_dict, mask, figure_path,*args):
    for key in list(brain_icc_dict.keys()):

        icc_data = brain_icc_dict[key].get_fdata()
        mask_data = mask.get_fdata()
        
        combined_data = icc_data * mask_data  # Assuming mask is binary (0s and 1s)
    
        combined_image = nib.Nifti1Image(combined_data, affine=brain_icc_dict[key].affine)
        nib.save(combined_image, os.path.join(figure_path, f'{key}_{args[0]}.nii'))
    
        #nib.save(brain_icc_dict[key], os.path.join(figure_path, f'{key}_{args[0]}.nii'))

```

## Create ICC maps for ses-1 ses-2 comparison of BETA VALUES

```python
# check if the lists of images have the same length
path_list_ses1=[]
path_list_ses2=[]

for sub in first_level_dirs:
    # if you want to compare beta values between 2 Session for all learning stage make sure you create a first level GLM with one regressor 
    session_images = ['beta_0001.nii','beta_0015.nii'] #L ses 1, L ses2   
    # Choose implicit baseline contrasts 
    
    for ses,img in enumerate(session_images):

        if not os.path.isfile(os.path.join(sub_root, sub, img)):
            print(sub_root, sub, img)
            continue
        else:
            # Create path to image
            #if you have 4 beta images or regressors for each stage change if condition to ses+1<5:
            if ses+1==1:
                path_list_ses1.append(os.path.join(sub_root, sub, img))
            #elif ses+1 >=5:
            else:
                path_list_ses2.append(os.path.join(sub_root, sub, img))


if len(path_list_ses1)==len(path_list_ses2):
    print('lists have same length')
```

```python
print(path_list_ses1)
print(path_list_ses2)
```

```python
# compute whole brain mask (this make take a while if the dataset is big)
from nilearn.masking import compute_multi_brain_mask
list_path = path_list_ses1 + path_list_ses2
mask = compute_multi_brain_mask(target_imgs = list_path)
whole_brain_mask_path = os.path.join(path, 'masks', 'multi_brain_mask.nii.gz')
nib.save(mask, whole_brain_mask_path)
```

## activity mask

```python
# Reminder: If NaN/missing values, icc implements a mean replacement of all column values. 
# If this is not preferred, handle missing/unbalanced cases beforehand.

brain_icc_dict_mask = brain_icc.voxelwise_icc(multisession_list = [path_list_ses1, path_list_ses2], mask = activity_mask, icc_type = "icc_3")
save_images(brain_icc_dict_mask, figure_path,'activity_mask_ses1_vs_ses2')



```

```python
# Create an interactive view of the 3D image to get coordinates for the correct cut
# click in the images to use the interactive view

# change threshold and coords for visualization 
view = plotting.view_img(brain_icc_dict_mask["est"], threshold=0.7)
#change coordinates to best show region of interests
coords = (32, -50, -14)

# Display the interactive plot
view
```

```python
# Plot the ICC map
fig = plotting.plot_stat_map(brain_icc_dict_mask["est"], threshold=0.5, title='ICC Map', display_mode='ortho', cut_coords=coords)

fig.savefig(os.path.join(figure_path,f'icc_map_activity_map_ses1_vs_ses2.png'))
# Show the plot
plotting.show()
```
